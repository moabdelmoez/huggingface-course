## What is Natural Language Processing?

NLP is a how to teach a computer to understand the human language, not only on the word level but the context of the context as well.

* Classifyinh whole sentences (ex, knowing the review is good or bad, detecting this mail is spam or not, check sentence grammer, ...)
* Classifying each word in a sentence (ex, identify person, country, ... "named entities")
* Generating text content
* Extracting an answer from a text
* Translating a text to another language, summerizing a text

> NLP is not limited to text, but also in computer vision like image captioning, and in speech recognition like transcript

Transoformer models are used to solve NLP, computer vision, and audio challenges to name a few.

Huggingface transformers library is the place to create or use other's models like from Google, Microsoft, ...

The model hub contains many pretrained models 

### `pipeline()`

The pipeline are a great, and easy way to use models (audio, computer vision, NLP, multimodal). It removes the complexity of the code, and offering a simple API. 

```python
!pip install datasets evaluate transformers[sentencepiece]
```

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
classifier("I love my family")
```

```python
# we can pass several sentences

classifier(["I have waiting for spiderman new series",
            "I hate last part from batman"])
```

Keep in mind there are 3 main steps when you pass some text to a pipeline;
1. the text is pre-processed, so the model can understand it
2. the pre-processed inputs are passed to the model
3. the predicitions are post-processed in a format we understand

For more information about `pipeline` - https://huggingface.co/docs/transformers/main_classes/pipelines

> Zero-shot learning allows a model to recognize what it hasn't seen before. CLIP "Contrastive Language-Image Pretraining" proposed by OpenAI has performed well in zero-shot learning.


```python
classifier = pipeline("zero-shot-classification")
classifier("this is a course about machine learning",
           candidate_labels=["education", "sport", "biology"])
```

```python
# text generation

generator = pipeline("text-generation")
generator("I love my mother a lot")
```

```python
generator("a family gathering is all about",
          num_return_sequences=2)
```

If you want to use a specific model from the hub, not just the default one.
https://huggingface.co/models?pipeline_tag=text-generation

```python
generator = pipeline("text-generation", model="distilgpt2")
generator("in this deep learing course, we will teach how to",
         max_length=30,
         num_return_sequences=2)
```

```python
# named entity recognition

ner = pipeline("ner", grouped_entities=True)
ner("my name is mostafa and I work at Intel. I live in Cario")
```

```python
from transformers.pipelines import question_answering
# question answering

question_answer = pipeline("question-answering")
question_answer(question="what is my name",
                context="I live in Maadi, and my friends called me sasa")
```

```python
# translation

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ar")
translator("here in egypt the weather is so amazing")
```

## Transfer Learning

Transfer Learning is the act of initializing a model with another model's weights.

Transfer Learning reduces the amount of compute, time and data used.

Transfer Learning transfers knowledge and bias as well.

Fine-tuning a model therefore has lower time, data, financial, and environmental costs.

> Pro Tip: Choose a pretrained model that is as close as possible to the task you have and fine-tune it. 



## Transformer architecture

transformers_blocks.svg

* Encoder - receives an input (such as text) and convert it into numerical representation. That means the main task for the encoder is to acquire understanding from the input

* Decoder - uses encoder's output and other inputs to generate outputs

We can use both of them in some tasks, and one of them in other task;

* encoder-only models for tasks like text classification, and NER (need to understand the inputs)
* decoder-only models for tasks like text generation
* encoder-decoder models for tasks like summerization, or translation (need to understand input to generate output)

### Attention Layers

The main goal for the attention layers is to tell the model to pay specific attention to certain words in the sentence.

transformers.svg

### Encoder models

> Bi-directional means the Transformer encoder reads the entire sequence of words at once (reads text from both directions). Therefore, it is considered bidirectional.

The encoder outputs the a numerical representation for each word used in the input.

These representation contain the the value of a word but contextualized (which means each word takes into account the words around it)

Models example; BERT

### Decoder models

The decoder outputs numerical representation from an intial sequence.

Decoder is using *masked self-attention*, which means words can only see the words on left side, the right side is hidden.

When should use decoder?
* uni-directional: access to their left (or right) context
* great at causal tasts; generating sequences
* Natural Language Generation

Models example; GPT-2, GPT-3

> auto-regressive models re-use their past outputs as inputs in the following steps

Seqence-to-Sequence models examples; BART, T5, ...
